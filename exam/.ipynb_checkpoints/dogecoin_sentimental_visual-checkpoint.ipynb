```python
import pandas as pd
from datetime import datetime, timedelta
import numpy as np
import matplotlib.pyplot as plt
import tqdm
```

```python
#Loading the CSV
dogecoin_senti_dirty_df = pd.read_csv('dogecoin_sentimental.csv', encoding='utf-8')
#Removing empty column
dogecoin_senti_df = dogecoin_senti_dirty_df.iloc[:, 1:]

#Rounding Publish date to hours
dogecoin_senti_df['Rounded Publish Date'] = [str[:-5]+'00:00' for str in dogecoin_senti_df['Publish Date']]
```

```python
#Add a datetime coulmn
dogecoin_senti_df['datetime'] = pd.to_datetime(dogecoin_senti_df['Rounded Publish Date'])
dogecoin_senti_df
```

```python
#Round down datetime to the hour
#dogecoin_senti_df['datetime'] = pd.to_datetime(dogecoin_senti_df['Publish Date'])
#dogecoin_senti_df['datetime_r'] = pd.to_datetime(dogecoin_senti_df['datetime']).dt.strftime('%Y-%M-%d %H:%M:%S')
dogecoin_grouping_df = dogecoin_senti_df[['Title', 'sentiment', 'datetime']]
#dogecoin_grouping_df['datetime_r_obj'] = dogecoin_grouping_df['datetime_r'].astype(str)
dogecoin_grouping_df.head(10)
```

```python
#create dataframe with range of data and hourly steps
rng = pd.date_range(start='2020/09/01', end ='2021/12/30',freq='H')
date_range_df = pd.DataFrame({'Date' : rng})
date_range_df
```

```python
#Function that groups by datetime, then by sentiment, sums negative and sums positive

PFL = []
for hour in tqdm.tqdm(date_range_df['Date']):
    try:
        dogecoin_grouped_dt_df = dogecoin_grouping_df.groupby('datetime')
        pos_g = dogecoin_grouped_dt_df.get_group(hour).groupby('sentiment').get_group(1)
        pos_freq = pos_g.shape[0]
        PFL.append(pos_freq)
        
    except:
        PFL.append(0)
        pass
print(PFL)
```

```python
NFL = []
for hour in tqdm.tqdm(date_range_df['Date']):
    try:
        dogecoin_grouped_dt_df = dogecoin_grouping_df.groupby('datetime')
        neg_g = dogecoin_grouped_dt_df.get_group(hour).groupby('sentiment').get_group(-1)
        neg_freq = neg_g.shape[0]
        NFL.append(neg_freq)
        
    except:
        NFL.append(0)
        pass
print(NFL)
```

```python
pos_f_df = pd.DataFrame(PFL)
pos_f_df.plot()
```

```python
neg_f_df = pd.DataFrame(NFL)
neg_f_df.plot()
```

```python
dogecoin_final_df = date_range_df
dogecoin_final_df['positive_freq'] = PFL
dogecoin_final_df['negative_freq'] = NFL
plt.plot(dogecoin_final_df.iloc[:, 0], dogecoin_final_df['positive_freq'], color ='tab:green') 
plt.plot(dogecoin_final_df.iloc[:, 0], dogecoin_final_df['negative_freq'], color ='tab:red')
plt.xlabel("Date")
plt.ylabel("Number of posts")
plt.show()
```

```python
dogecoin_final_df.to_csv('dogecoin_sentiment_analysis.csv')
```
